```python
from scipy.optimize import curve_fit

def exponential(x, a, b):
    return a*np.exp(b*x)

```

```python
zipcs=list(data['recentmod'].sort_values(by="COVID_CASE_RATE_WEEK4", ascending=False).MODIFIED_ZCTA.head(25))
df=data['recentmod'].groupby("MODIFIED_ZCTA").sum()
ys=[]
x=[1, 2, 3, 4]

for zipc in zipcs:
    w1=int(df[df.index==zipc].COVID_CASE_COUNT_WEEK1)
    w2=int(df[df.index==zipc].COVID_CASE_COUNT_WEEK2)
    w3=int(df[df.index==zipc].COVID_CASE_COUNT_WEEK3)
    w4=int(df[df.index==zipc].COVID_CASE_COUNT_WEEK4)
    ys.append([w4, w3, w2, w1])
```

```python
i1=[0, 1, 2, 3, 4]
i2=[0, 1, 2, 3, 4]

fig, ax= plt.subplots(5, 5, figsize=(25, 25), constrained_layout=True)

for i1val in i1:
    for i2val in i2:
        ival=5*i1val+i2val
        ax[i1val][i2val].bar(x, ys[ival])
        ax[i1val][i2val].set_title(zipcs[ival])

```
